### Reading data from multiple sources
Read an HTML table from wikipedia

Full article [here](https://pbpython.com/pandas-html-table.html)

```python
import pandas as pd
import numpy as np
```

```python
table_MN = pd.read_html('https://en.wikipedia.org/wiki/Minnesota', match='Election results from statewide races')
df = table_MN[0]
```

```python
df
```

```python
df = df.replace({'%': ''}, regex=True)
df[['GOP', 'DFL', 'Others']] = df[['GOP', 'DFL',
                                   'Others']].apply(pd.to_numeric)
df.info()
```
