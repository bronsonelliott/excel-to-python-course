### Pandas merge is more powerful than vlookup

```python
import pandas as pd
from pathlib import Path
```

```python
shipping_file = Path.cwd() / 'data' / 'raw' / 'shipping_tables.xlsx'
```

```python
df_sales = pd.read_excel(shipping_file, sheet_name='sales', usecols='A:E')
df_shipping = pd.read_excel(shipping_file, sheet_name='shipping_rates')
```

```python
df_sales
```

```python
df_shipping
```

```python
results = pd.merge(df_sales, df_shipping, how='left')
results
```
